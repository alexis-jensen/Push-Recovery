```python
push_data = pd.read_csv('PushData.csv',sep=";")[['MaxForce', 'PerturbationDuration','Final_CoM_y','angle','StepIndex','Impulse']]
push_data = push_data.loc[push_data['angle']==90]
push_data2 = push_data.loc[push_data['StepIndex']==2]
push_data2=push_data2[['MaxForce', 'PerturbationDuration','Final_CoM_y','Impulse']]
push_data = push_data.loc[push_data['StepIndex']==0]
push_data=push_data[['MaxForce', 'PerturbationDuration','Final_CoM_y','Impulse']]

print(len(push_data))

push_data.head()
push_data2.head()
```

## Fonction evaluant la stabilité du CoM en fonction des poids kd,kp

```python
def weight_impact_function(param,force,time):
   
    #kds = kd1,kd2,kd3,kd4,kd5,kd6,kd7,kd8,kd9,kd10,kd11,kd12,kd13,kd14,kd15,kd16,kd17,kd18
    #kps = kp1,kp2,kp3,kp4,kp5,kp6,kp7,kp8,kp9,kp10,kp11,kp12,kp13,kp14,kp15,kp16,kp17,kp18
    
    #print(param)
    
    for i in {1,19,33,35}:
        param[i]=0
    
    
    f = open("input.txt", "w")
    f.write(('\n'.join(str(param[i]) + " "+ str(param[i+18]) for i in range(18))).replace("[","").replace("]",""))
    f.write("\n" + str(force) + " " + str(time) +"\n")
    f.close()

    subprocess.check_call(['../../bin/App_BulletExampleBrowser_vs2010_x64_debug.exe'])

    donnee_1 = pd.read_csv('output.txt', sep=" ")
    
    result = donnee_1["Z"].iat[-1]
    return result


def all_particles(input_param,force,time):
    result =[0] * particles
    
    
    for j in range(len(push_data)):
        current_test = push_data.iloc[random.randint(0,len(push_data)-1)].to_numpy()
        #current_test = push_data.iloc[j].to_numpy()
        print(current_test)
        for i in range(particles):
            temp = -(abs(current_test[2])+weight_impact_function(input_param[i],current_test[0],current_test[1]))
            result[i] = result[i] + temp
            print("Diff "+ str(j) + "/" + str(len(push_data)) +" : " + str(-abs(current_test[2])) + " - " + str(temp+abs(current_test[2])) + " -> " + str(temp) +"\n")
            dataa.append([current_test[0],current_test[1],current_test[2],temp,current_test[3]])
            
    for j in range(len(push_data2)):
        current_test = push_data2.iloc[random.randint(0,len(push_data2)-1)].to_numpy()
        #current_test = push_data.iloc[j].to_numpy()
        print(current_test)
        for i in range(particles):
            temp = -(abs(current_test[2])+weight_impact_function(input_param[i],current_test[0],current_test[1]))
            result[i] = result[i] + temp
            print("Diff "+ str(j) + "/" + str(len(push_data2)) +" : " + str(-abs(current_test[2])) + " - " + str(temp+abs(current_test[2])) + " -> " + str(temp) +"\n")
            dataa.append([current_test[0],current_test[1],current_test[2],temp,current_test[3]])
            
            
    print(result)
    return result
        
        
```

## Optimiser

```python
dataa=[]

kwargs = { "kd2":0, "kd3":-900, "kd4":-900, "kd5":-20, "kd6":-20, "kd7":-900, "kd8":-900, "kd9":-20, "kd10":-20, "kd11":-20, "kd12":-20, "kd13":-1200, "kd14":-1200, "kd15":-20, "kd16":-10, "kd17":-20, "kd18":-10,"kp2":0,"kp3":-10, "kp4":-10, "kp5":-1, "kp6":-1, "kp7":-10, "kp8":-10, "kp9":-1.5, "kp10":-1.5, "kp11":-1.5, "kp12":-1.5, "kp13":-20, "kp14":-20, "kp15":-1, "kp16":0, "kp17":-1, "kp18":0,"force":36,"time":0.55}
kwargs={"force":36,"time":0.55}
max_bound = np.array([-500,0,-300, -500, -500, -500, -300, -500, -10, -10, -10, -10, -1000, -1000, -50, -50, -50, -50,-100,0,-10, -20, -20, -20, -20, -20, -1, -1, -1, -1, -20, -20, -5, 0, -5, 0])
min_bound = np.array([-500,0,-300, -500, -500, -500, -300, -500, -10, -10, -10, -10, -1000, -1000, -50, -50, -50, -50,-100,0,-10, -20, -20, -20, -20, -20, -1, -1, -1, -1, -20, -20, -5, 0, -5, 0])
bounds = (min_bound, max_bound)
options = {'c1': 0.5, 'c2': 0.3, 'w':0.9}
optimizer = ps.single.GlobalBestPSO(n_particles=particles, dimensions=36, options=options, bounds=bounds)
cost, pos = optimizer.optimize(all_particles, **kwargs,iters=iterations)
```

```python
from pyswarms.utils.plotters.formatters import Mesher


# Initialize mesher with sphere function
m = Mesher(func=fx.sphere)


# Obtain a position-fitness matrix using the Mesher.compute_history_3d()
# method. It requires a cost history obtainable from the optimizer class
pos_history_3d = m.compute_history_3d(optimizer.pos_history)
result= [[[0,0] for i in range(len(pos_history_3d[0])) ]for j in range(len(pos_history_3d)) ]
for i in range(len(pos_history_3d)):
    for j in range(len(pos_history_3d[0])):
        result[i][j] = pos_history_3d[i][j][0:2]
print(result)

pos_history_3d = result

# Make a designer and set the x,y,z limits to (-1,1), (-1,1) and (-0.1,1) respectively
from pyswarms.utils.plotters.formatters import Designer
d = Designer(limits=[(-1,1), (-1,1), (-0.1,1)], label=['x-axis', 'y-axis', 'z-axis'])

```

```python
liste= pos_history_3d

itera=10
partic=5

interp=30

x = []
y = []
for i in range(itera):
        for j in range(partic):
            x.append(liste[i][j][0])
            y.append(liste[i][j][1])   
            
            

fig, ax = plt.subplots()
points, = ax.plot(x, y, marker='o', linestyle='None')

 
            
progress = 1

def animate(frame_num):
    plt.cla()
    ax.set_xlabel('Valeur de Kp')
    ax.set_ylabel('Valeur de Kd')
    ax.set_title('Swarm Optimisation')
    
    
    ax.set_xlim(-1300,-1100)
    ax.set_ylim(-25,-15)
    
    x=[]
    y=[]
    for j in range(partic):
        x.append(liste[floor(frame_num/interp)][j][0])
        y.append(liste[floor(frame_num/interp)][j][1])
        
        if(floor(frame_num/interp) < itera):
            x[j] = [(1-(frame_num%interp * 1/interp)) * x[j] + frame_num%interp * 1/interp * liste[floor(frame_num/interp)+1][j][0]]
            y[j] = [(1-(frame_num%interp * 1/interp)) * y[j] + frame_num%interp * 1/interp * liste[floor(frame_num/interp)+1][j][1]]

    points, = ax.plot(x, y, marker='o', linestyle='None')
    
    points.set_data((x,y))
    
    return points

anim = FuncAnimation(fig, animate, frames=itera*interp, interval=5, repeat=False)
plt.show()
print(liste)
```
